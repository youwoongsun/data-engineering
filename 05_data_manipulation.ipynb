## **5. Pandas 데이터 조작 실습**

### **패키지 임포트 및 DataFrame 객체 생성**
가장 먼저 데이터 조작에 유용한 파이썬 라이브러리인 Pandas 패키지를 `pd`라는 이름으로 임포트합니다. 
이후 `.csv` 파일을 인덱싱된 2차원 배열 구조인 `DataFrame` 객체로 불러옵니다.

```python
import pandas as pd
import numpy as np

# 데이터 로드 (csv 파일 읽기)
df = pd.read_csv('./data/AirQualityUCI_refined.csv')

# 데이터의 첫 5행을 확인하여 올바르게 로드되었는지 확인
df.head()
```

### **Pandas 객체의 주요 속성 확인 (Main Attributes)**
Pandas 객체가 기본적으로 제공하는 주요 속성을 확인합니다.
* `.index`: 행의 인덱스 값을 반환합니다.
* `.columns`: DataFrame의 열 레이블(Column 명)을 반환합니다.
* `.values`: 데이터의 값들을 NumPy 2차원 배열(ndarray) 형태로 반환합니다.

```python
# 행 인덱스 반환
df.index
```

```python
# 열 레이블 반환
df.columns
```

```python
# 값들을 NumPy 배열로 반환 (너무 길어지므로 첫 2행만 출력)
df.values
```

### **Series 객체 조작 (Dictionary Style)**
DataFrame은 인덱스를 공유하는 1차원 배열인 `Series` 객체들의 집합과 같습니다. 
파이썬의 Dictionary 구조와 유사하게 열 이름(Key)을 통해 특정 Series를 선택하거나, 새로운 데이터를 추가(Reassignment)할 수 있습니다.

```python
# 'CO(GT)' 열을 하나의 Series 객체로 선택
co_series = df['CO(GT)']
print(type(co_series))
print("CO(GT) Series 첫 3개 값:\n", co_series.head(3))

# Dictionary 스타일의 새로운 데이터 열 추가
# 기존 CO(GT) 값에 2를 곱한 새로운 열 생성
df['CO_Double'] = df['CO(GT)'] * 2

print("\n새로운 열 추가 확인:")
df[['CO(GT)', 'CO_Double']].head()
```

### **Indexers: 명시적 인덱스(loc)와 암묵적 인덱스(iloc)**
데이터를 추출하거나 슬라이싱할 때 활용되는 인덱서입니다.
* `iloc` (Integer-based Locating): 암묵적인 정수 인덱스(0, 1, 2...)를 참조하여 슬라이싱을 수행합니다.
* `loc` (Label-based Locating): 사용자가 명시적으로 정의한 인덱스 레이블을 참조하여 슬라이싱을 수행합니다.

```python
# 암묵적 인덱싱: 처음 3개의 행과 처음 4개의 열을 정수로 슬라이싱
df.iloc[0:3, 0:4]
```

```python
# 'Datetime'을 행의 명시적 인덱스로 설정
df_time_idx = df.set_index('Datetime')

# 레이블 기반 인덱싱 실습: 특정 시간(문자열 레이블)부터 시작하여 특정 열들만 슬라이싱
df_time_idx.loc['2004-03-10':'2004-03-15', ['CO(GT)', 'NOx(GT)', 'NO2(GT)']]
```

### **마스킹 (Masking) 연산**
조건에 부합하는 특정 데이터만 걸러내기 위해 배열 스타일의 마스킹(Boolean Indexing) 연산을 사용할 수 있습니다.

```python
# 조건: 이산화질소 센서 반응값 'PT08.S4(NO2)'가 1500을 초과하는 경우
mask_condition = df['PT08.S4(NO2)'] > 1500.0

# 마스크를 씌워 필터링된 새로운 DataFrame 생성
high_no2_df = df[mask_condition]

print(f"조건을 만족하는 관측치 개수: {len(high_no2_df)}개")
high_no2_df[['Datetime', 'PT08.S4(NO2)']].head()
```

### **전치 행렬 (Transposed Matrix)**
필요에 따라 데이터셋을 재구조화할 때, 행과 열의 위치를 바꾸는 전치 행렬(`.T`)을 구할 수 있습니다.

```python
# 데이터의 첫 3행을 추출한 후 행과 열을 전치(Transpose)
df_transposed = df.head(3).T
print("전치 행렬 결과:\n", df_transposed)
```

### **탐색적 데이터 분석 (Exploratory Data Analysis, EDA)**
데이터를 불러온 후에는 데이터의 형태, 속성, 통계량 등을 파악하여 데이터에 대한 이해도를 높여야 합니다.
* `.info()`: 데이터프레임의 행/열 개수, 각 열의 데이터 타입, 결측치가 아닌(Non-Null) 데이터의 개수 등 전반적인 정보를 출력합니다.
* `.describe()`: 수치형 데이터의 기술 통계량(개수, 평균, 표준편차, 최솟값, 4분위수, 최댓값)을 요약하여 보여줍니다.
* `.isnull().sum()`: 각 열마다 결측치(NaN)가 몇 개 존재하는지 확인합니다.

```python
# 데이터 프레임의 기본 정보 확인
df.info()
```

```python
# 결측치(Missing Value) 개수 확인
print(df.isnull().sum())
```

```python
# 수치형 데이터의 기술 통계량 확인
df.describe()
```

### **EDA 심화: 변수 간의 상관관계 확인**
연속형 수치 데이터로 이루어진 경우, 변수들 간의 선형적 관계를 파악하기 위해 상관계수(Correlation Coefficient)를 확인할 수 있습니다.
* `.corr()`: 수치형 열 사이의 피어슨 상관계수(-1부터 1 사이의 값)를 계산합니다.

```python
# 'Datetime'과 같은 문자열/시간 데이터는 상관계수 계산에서 제외하기 위해 수치형 데이터만 선택
numeric_df = df.select_dtypes(include=[np.number])

# 변수 간의 상관계수 행렬 출력 (예: CO(GT)와 PT08.S1(CO)의 상관관계 확인)
corr_matrix = numeric_df.corr()

corr_matrix
```
